```
pacientes = []

def cadastrar_paciente():
    try:
        nome = input("Digite o nome do paciente: ").strip()
        idade = int(input('Digite a idade do paciente: '))
        telefone = input('Digite o telefone do paciente: ').strip()
        paciente = {"nome": nome, "idade": idade, "telefone": telefone}
        pacientes.append(paciente)
        print(f'Paciente {nome} cadastrado com sucesso!')
    except ValueError:
        print('Idade inválida. Por favor, insira um número inteiro para a idade.')

def idade_media():
    if len(pacientes) == 0:
        return 0
    total_idade = sum(p['idade'] for p in pacientes)
    return total_idade / len(pacientes)

def buscar_paciente():
    nome_buscar = input('Digite o nome do paciente para busca: ').strip().lower()
    resultados = [p for p in pacientes if p['nome'].lower() == nome_buscar]
    if resultados:
        for p in resultados:
            print(f'Nome: {p["nome"]}, Idade: {p["idade"]}, Telefone: {p["telefone"]}')
    else:
        print('Paciente não encontrado.')

def exibir_pacientes():
    if not pacientes:
        print('Nenhum paciente cadastrado.')
    else:
        print('Pacientes cadastrados:')
        for i, p in enumerate(pacientes, 1):
            print(f"{i}. Nome: {p['nome']}, Idade: {p['idade']}, Telefone: {p['telefone']}")

def extremos_idade():
    if not pacientes:
        return None, None
    mais_novo = min(pacientes, key=lambda p: p['idade'])
    mais_velho = max(pacientes, key=lambda p: p['idade'])
    return mais_novo, mais_velho

def menu():
    while True:
        print('\nMenu:')
        print('1. Cadastrar paciente')
        print("2. Exibir total de pacientes e estatísticas")
        print("3. Buscar paciente por nome")
        print("4. Exibir todos os pacientes")
        print("5. Sair")
        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            cadastrar_paciente()
        elif escolha == '2':
            total = len(pacientes)
            media = idade_media()
            mais_novo, mais_velho = extremos_idade()
            print(f"\nTotal de pacientes cadastrados: {total}")
            print(f"Idade média dos pacientes: {media:.2f}")
            if mais_novo and mais_velho:
                print(f"Paciente mais novo: {mais_novo['nome']}, idade: {mais_novo['idade']}")
                print(f"Paciente mais velho: {mais_velho['nome']}, idade: {mais_velho['idade']}")
            else:
                print("Não há pacientes para exibir estatísticas de idade.")
        elif escolha == '3':
            buscar_paciente()
        elif escolha == '4':
            exibir_pacientes()
        elif escolha == '5':
            print("Saindo do programa.")
            break
        else:
            print("Opção inválida. Tente novamente.")

menu()
```
